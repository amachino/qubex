# State and pulse tomography

This notebook compares standard and DRAG half-pi pulses using state
tomography and pulse tomography. The same pattern works for any waveform
or `PulseSchedule` you want to reconstruct.

## 1. Create an `Experiment`

```python
import qubex as qx

exp = qx.Experiment(
    system_id="YOUR_SYSTEM_ID",
    qubits=[0],
    # config_dir="/path/to/qubex-config/config",
    # params_dir="/path/to/qubex-config/params/YOUR_SYSTEM_ID",
)
```

```python
Q0 = exp.qubit_labels[0]

print("target qubit:", Q0)
```

## 2. Connect and calibrate the half-pi pulses

Tomography is much easier to interpret once the basic pulse amplitudes are
already close to the desired rotations.

```python
exp.connect()

rabi_result = exp.obtain_rabi_params(targets=Q0, n_shots=512, plot=True)
hpi_result = exp.calibrate_hpi_pulse(targets=Q0, n_shots=512, plot=True)
drag_hpi_result = exp.calibrate_drag_hpi_pulse(
    targets=Q0,
    n_iterations=2,
    n_shots=512,
    plot=True,
)
```

## 3. Inspect the cached pulses

```python
exp.hpi_pulse[Q0].plot()
exp.drag_hpi_pulse[Q0].plot()
```

## 4. Run state tomography

Compare the state prepared by the default half-pi pulse and the DRAG
version.

```python
state_default = exp.state_tomography(
    {Q0: exp.hpi_pulse[Q0]},
    initial_state={Q0: "0"},
    n_shots=1024,
    plot=True,
)

state_drag = exp.state_tomography(
    {Q0: exp.drag_hpi_pulse[Q0]},
    initial_state={Q0: "0"},
    n_shots=1024,
    plot=True,
)
```

## 5. Run pulse tomography

Pulse tomography reconstructs the effective rotation generated by the same
pulse primitives.

```python
pulse_default = exp.pulse_tomography(
    {Q0: exp.hpi_pulse[Q0]},
    n_shots=1024,
    plot=True,
)

pulse_drag = exp.pulse_tomography(
    {Q0: exp.drag_hpi_pulse[Q0]},
    n_shots=1024,
    plot=True,
)
```
