```python
import qubex as qx

exp = qx.Experiment(
    chip_id="xxx",
    muxes=[0],
)
```

```python
exp.connect()
```

```python
exp.check_waveform()
```

```python
exp.obtain_rabi_params()
```

```python
exp.calibrate_pi_pulse()
exp.calibrate_drag_pi_pulse()
exp.calib_note.save()
```

```python
exp.repeat_sequence(exp.pi_pulse)
exp.repeat_sequence(exp.drag_pi_pulse)
```

```python
target = exp.qubit_labels[0]

pi = exp.pi_pulse[target]
pi.plot()

drag_pi = exp.drag_pi_pulse[target]
drag_pi.plot()
```
