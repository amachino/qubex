# MeasurementResult NetCDF Codec

This notebook demonstrates how to save and load `MeasurementResult` using `MeasurementResultNetCDFCodec`.

```python
from pathlib import Path

import numpy as np

from qubex.measurement.models import (
    MeasurementResult,
    MeasurementResultNetCDFCodec,
)
```

```python
result = MeasurementResult(
    mode="single",
    data={
        "Q62": [np.array([[1 + 2j, 3 + 4j]])],
        "Q63": [np.array([[5 + 6j, 7 + 8j]])],
    },
    device_config={"backend": "quel"},
    measurement_config={"mode": "single", "shots": 1024},
)

codec = MeasurementResultNetCDFCodec()
```

```python
output_path = Path(
    "docs/examples/measurement/.rawdata/example_measurement_result_codec.nc"
)
output_path.parent.mkdir(parents=True, exist_ok=True)
saved_path = codec.save(result, output_path)
print(saved_path)
```

```python
restored = codec.load(saved_path)
print("mode:", restored.mode)
print("targets:", list(restored.data.keys()))
print("measurement_config:", restored.measurement_config)
```

```python
for target, captures in restored.data.items():
    for capture_idx, arr in enumerate(captures):
        print(target, capture_idx, arr.shape, arr.dtype)
```
