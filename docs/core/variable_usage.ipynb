# Variable Usage

`qubex.core.variable.Variable` allows you to define symbolic expressions and resolve them later with specific values.

```python
from qubex.core.variable import Variable
```

## Basic Usage

```python
# Define a variable with an expression string
var = Variable("2 / a * x + b")
var
```

```python
# Resolve the variable by providing a value map
result = var.resolve(
    {
        "a": 10,
        "x": 5,
        "b": 3,
    }
)
print(f"Result: {result}")  # 2 / 10 * 5 + 3 = 4.0
```

## Using Mathematical Functions

You can use standard mathematical functions supported by `numpy`.

```python
import numpy as np

gaussian = Variable("amp * exp(-(x - mu)**2 / (2 * sigma**2))")

params = {"amp": 1.0, "mu": 0.0, "sigma": 1.0, "x": np.linspace(-3, 3, 5)}

values = gaussian.resolve(params)
print("Computed values:", values)
```
